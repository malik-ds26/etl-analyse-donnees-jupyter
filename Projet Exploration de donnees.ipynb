###Fonction pour les valeurs manquantes :

```python

def handle_missing_values(df, strategy="mean"):

    df_cleaned = df.copy()

    if strategy == "mean":
        df_cleaned.fillna(df_cleaned.mean(numeric_only=True), inplace=True)
    elif strategy == "median":
        df_cleaned.fillna(df_cleaned.median(numeric_only=True), inplace=True)
    elif strategy == "mode":
        df_cleaned.fillna(df_cleaned.mode().iloc[0], inplace=True)
    elif strategy == "drop":
        df_cleaned.dropna(inplace=True)
    else:
        raise ValueError("Invalid strategy. Choose from 'mean', 'median', 'mode', or 'drop'.")

    return df_cleaned
```

```python
data_NO2_cleaned = handle_missing_values(data_NO2, strategy="mean")
data_NO2_cleaned
```

###Fonction Harmoniser les formats de date :

```python


def harmonize_data_formats(df, date_columns=None, lat_col=None, lon_col=None):

    df_harmonized = df.copy()


    if date_columns:
        for col in date_columns:
            df_harmonized[col] = pd.to_datetime(df_harmonized[col], errors='coerce', dayfirst=True)


    if lat_col and lon_col:
        df_harmonized[lat_col] = pd.to_numeric(df_harmonized[lat_col], errors='coerce')
        df_harmonized[lon_col] = pd.to_numeric(df_harmonized[lon_col], errors='coerce')

    return df_harmonized
```

###Fonction pour Normaliser vos données :

```python


def normalize_data(df, columns=None, method="minmax"):

    df_normalized = df.copy()


    if columns is None:
        columns = df_normalized.select_dtypes(include=["number"]).columns


    if method == "minmax":
        scaler = MinMaxScaler()
    elif method == "zscore":
        scaler = StandardScaler()
    else:
        raise ValueError("Invalid method. Choose 'minmax' or 'zscore'.")


    df_normalized[columns] = scaler.fit_transform(df_normalized[columns])

    return df_normalized
```

###Fonction pour gérer les incohérences potentielles :

```python
def handle_data_inconsistencies(df, value_ranges=None, categorical_columns=None):

    df_cleaned = df.copy()


    if value_ranges:
        for col, (min_val, max_val) in value_ranges.items():
            if col in df_cleaned.columns:
                df_cleaned[col] = df_cleaned[col].apply(lambda x: x if min_val <= x <= max_val else None)

    if categorical_columns:
        for col in categorical_columns:
            if col in df_cleaned.columns:

                mode_value = df_cleaned[col].mode()[0]
                df_cleaned[col] = df_cleaned[col].apply(lambda x: x if x in df_cleaned[col].unique() else mode_value)


    df_cleaned.dropna(inplace=True)

    return df_cleaned
```

```python
def handle_outliers(df, value_ranges):

    df_cleaned = df.copy()


    for col, (min_val, max_val) in value_ranges.items():
        if col in df_cleaned.columns:
            df_cleaned[col] = df_cleaned[col].apply(
                lambda x: x if min_val <= x <= max_val else None
            )


    df_cleaned.dropna(inplace=True)

    return df_cleaned
```

###Integration des données météorologiques :

```python
def obtenir_donnees_meteo(lat, lon, api_key):

    url = f"http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units=metric"

    try:
        response = requests.get(url)
        data = response.json()

        if response.status_code == 200:

            meteo = {
                'temp': data['main']['temp'],  # Température en °C
                'humidity': data['main']['humidity'],  # Humidité
                'pressure': data['main']['pressure'],  # Pression atmosphérique
                'wind_speed': data['wind']['speed'],  # Vitesse du vent
                'weather': data['weather'][0]['description']  # Description météo
            }
            return meteo
        else:
            print(f"Erreur {response.status_code}: Impossible de récupérer les données météo.")
            return None
    except Exception as e:
        print(f"Erreur lors de la récupération des données météo : {e}")
        return None
```
